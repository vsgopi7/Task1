```python
import pandas as pd


file_path = r"C:\Users\venka\OneDrive\Desktop\aadi foundation\Task1\food_coded.csv"
df = pd.read_csv(file_path)

print("First few rows of the dataset:")
print(df.head())


print("\nDataset info:")
print(df.info())


print("\nMissing values per column:")
print(df.isnull().sum())

for column in df.select_dtypes(include=['float64', 'int64']).columns:
    df[column].fillna(df[column].mean(), inplace=True)


for column in df.select_dtypes(include=['object']).columns:
    df[column].fillna(df[column].mode()[0], inplace=True)


df.drop_duplicates(inplace=True)


df.columns = [col.lower().replace(' ', '_') for col in df.columns]


cleaned_file_path = r"C:\Users\venka\OneDrive\Desktop\aadi foundation\Task1\food_coded_cleaned.csv"
df.to_csv(cleaned_file_path, index=False)

cleaned_df = pd.read_csv(cleaned_file_path)


print("First few rows of the cleaned dataset:")
print(cleaned_df.head())

print("\nDataset info:")
print(cleaned_df.info())

print("\nSummary statistics:")
print(cleaned_df.describe(include='all'))


```
